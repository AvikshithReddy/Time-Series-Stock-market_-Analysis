### **Moving Average**

Moving Average (MA) is widely used in technical analysis to smooth out the price by taking out the "noise" from random short-term price changes. Since it is based on historical prices, it is a lagging or trend-following indicator.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

for df in [shop, jd, alibaba]:
    df['MA50'] = df['Open'].rolling(window=50).mean()
    df['MA200'] = df['Open'].rolling(window=200).mean()
```

```python
# Shopify
plt.figure(figsize=(16, 6))
plt.plot(shop['Date'], shop['Open'], label='Shopify Open')
plt.plot(shop['Date'], shop['MA50'], label='Shopify MA50')
plt.plot(shop['Date'], shop['MA200'], label='Shopify MA200')
plt.title('Shopify - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# JD.com
plt.figure(figsize=(16, 6))
plt.plot(jd['Date'], jd['Open'], label='JD Open')
plt.plot(jd['Date'], jd['MA50'], label='JD MA50')
plt.plot(jd['Date'], jd['MA200'], label='JD MA200')
plt.title('JD.com - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Alibaba
plt.figure(figsize=(16, 6))
plt.plot(alibaba['Date'], alibaba['Open'], label='Alibaba Open')
plt.plot(alibaba['Date'], alibaba['MA50'], label='Alibaba MA50')
plt.plot(alibaba['Date'], alibaba['MA200'], label='Alibaba MA200')
plt.title('Alibaba - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## Basic Financial Analysis

By looking at a daily percentage change, it can be concluded stock's volatility and how risky it is. The formula is simple: more volatility = more risk.

Calculates daily percentage returns for each stock based on the previous dayâ€™s closing price.

```python
import matplotlib.pyplot as plt

# Step 1: Calculate Daily Returns
shop['Returns'] = shop['Close'].pct_change()
jd['Returns'] = jd['Close'].pct_change()
alibaba['Returns'] = alibaba['Close'].pct_change()

# Step 2: Drop NaNs (first row will be NaN after pct_change)
shop = shop.dropna(subset=['Returns'])
jd = jd.dropna(subset=['Returns'])
alibaba = alibaba.dropna(subset=['Returns'])

# Step 3: Plot Histograms separately for clarity
plt.figure(figsize=(18, 5))

# Shopify
plt.subplot(1, 3, 1)
plt.hist(shop['Returns'], bins=50, alpha=0.7, color='blue')
plt.title('Shopify Daily Returns')
plt.xlabel('Return')
plt.ylabel('Frequency')

# JD.com
plt.subplot(1, 3, 2)
plt.hist(jd['Returns'], bins=50, alpha=0.7, color='orange')
plt.title('JD.com Daily Returns')
plt.xlabel('Return')

# Alibaba
plt.subplot(1, 3, 3)
plt.hist(alibaba['Returns'], bins=50, alpha=0.7, color='green')
plt.title('Alibaba Daily Returns')
plt.xlabel('Return')

plt.tight_layout()
plt.show()
```

Shopify
The distribution is tall and narrow, centered around 0.

Has a few extreme outliers, both positive and negative (Â±15â€“20%).

Indicates moderate volatility with occasional sharp movements.

2. JD.com
Similar shape to Shopify but more concentrated around 0.

Fewer extreme returns, suggesting lower volatility.

Reflects a relatively stable return pattern.

3. Alibaba
Distribution is also centered around 0, slightly wider than JD.com.

Shows more frequent small returns but also a few spikes.

Indicates moderate volatility, but less than Shopify.

By plotting all 3 on the graph, let us compare them back to back.

```python
import matplotlib.pyplot as plt

# Ensure returns are calculated and cleaned
shop['Returns'] = shop['Close'].pct_change()
jd['Returns'] = jd['Close'].pct_change()
alibaba['Returns'] = alibaba['Close'].pct_change()

# Align data on Date and drop NaNs from all
returns_df = pd.concat([
    shop[['Date', 'Returns']].set_index('Date').rename(columns={'Returns': 'Shopify'}),
    jd[['Date', 'Returns']].set_index('Date').rename(columns={'Returns': 'JD.com'}),
    alibaba[['Date', 'Returns']].set_index('Date').rename(columns={'Returns': 'Alibaba'})
], axis=1).dropna()

# Plot all returns together
plt.figure(figsize=(16, 6))
plt.plot(returns_df.index, returns_df['Shopify'], label='Shopify', alpha=0.7)
plt.plot(returns_df.index, returns_df['JD.com'], label='JD.com', alpha=0.7)
plt.plot(returns_df.index, returns_df['Alibaba'], label='Alibaba', alpha=0.7)
plt.axhline(0, linestyle='--', linewidth=1, color='black', alpha=0.5)
plt.title('Daily Returns Comparison')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Itâ€™s the standard deviation of daily returns over a moving window (e.g., 7 days). below----->

```python
import matplotlib.pyplot as plt

# Step 1: Ensure daily returns are calculated
shop['Returns'] = shop['Close'].pct_change()
jd['Returns'] = jd['Close'].pct_change()
alibaba['Returns'] = alibaba['Close'].pct_change()

# Step 2: Compute 7-day rolling standard deviation
shop['Volatility'] = shop['Returns'].rolling(window=7).std()
jd['Volatility'] = jd['Returns'].rolling(window=7).std()
alibaba['Volatility'] = alibaba['Returns'].rolling(window=7).std()

# Step 3: Align and plot
vol_df = pd.concat([
    shop[['Date', 'Volatility']].set_index('Date').rename(columns={'Volatility': 'Shopify'}),
    jd[['Date', 'Volatility']].set_index('Date').rename(columns={'Volatility': 'JD.com'}),
    alibaba[['Date', 'Volatility']].set_index('Date').rename(columns={'Volatility': 'Alibaba'})
], axis=1).dropna()

# Step 4: Plot
plt.figure(figsize=(16, 6))
plt.plot(vol_df.index, vol_df['Shopify'], label='Shopify', alpha=0.8)
plt.plot(vol_df.index, vol_df['JD.com'], label='JD.com', alpha=0.8)
plt.plot(vol_df.index, vol_df['Alibaba'], label='Alibaba', alpha=0.8)

plt.title('7-Day Rolling Volatility (Standard Deviation of Daily Returns)')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Shopify shows the highest volatility spikes, peaking above 10%, indicating it is the most sensitive and reactive stock among the three â€” especially during market turbulence.

JD.com and Alibaba exhibit more stable volatility, generally staying below 5%, with Alibaba being the least volatile overall, making them relatively safer from a risk perspective.
